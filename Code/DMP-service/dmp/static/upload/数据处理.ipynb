## 使用Datetime模块

```python
import datetime
```

```python
def toDate(t):
    date = str(datetime.datetime.fromtimestamp(t).date())
    return date
```

```python
def toHour(t):
    hour = datetime.datetime.fromtimestamp(t).hour
    return hour
```

```python
toDate(1512293792)
```

```python
toHour(1512293792)
```

```python
data_1
```

```python
data_1['date'] = data_1['time'].map(toDate)
data_1['hour'] = data_1['time'].map(toHour)
```
